### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
import pandas as pd
import numpy as np

# 1. Load data
fifa_df = pd.read_csv(
    "https://docs.google.com/spreadsheets/d/e/2PACX-1vT3D_x_4DS6d51LKJ7ze1sxT5WpV5uiSVOFYHLwBiGru6vFyVv5h5-83AwFjxWYiWfCDjDAaarHAV-k/pub?gid=0&single=true&output=csv"
)
````

````python
print(fifa_df.columns)
print("\n")
print(fifa_df.head())
````

````python
print(fifa_df[['Team', 'Passes', 'Passes Completed', 'Attempted Line Breaks', 'Completed Line Breaks']].info())
````

````python
# 2. Define z_normalization function that will be applied to each group
def z_normalization(group: pd.DataFrame, cols_to_perform: list) -> pd.DataFrame:
    """
    Takes a DataFrame 'group' (one team's rows) and a list of columns.
    Returns the same group with new columns <col>_z containing z-scores.
    If std == 0 for a column, z-values are set to 0 to avoid division by zero.
    """
    # operate on a copy to avoid SettingWithCopyWarning
    group = group.copy()

    for col in cols_to_perform:
        # convert to numeric (coerce errors to NaN), helps if there are stray strings
        values = pd.to_numeric(group[col], errors='coerce')

        mu = values.mean()                  # group mean
        sigma = values.std(ddof=0)          # population std (ddof=0). Use ddof=1 if you want sample std.
        if pd.isna(sigma) or sigma == 0:
            # all values same or insufficient data -> set z to 0
            z = (values - mu) * 0
        else:
            z = (values - mu) / sigma

        # add new column to group, round if desired (here we keep raw)
        group[f"{col}_z"] = z.round(6)     # rounding just for neatness

    return group
````

````python
# 3. Columns on which to perform Z-normalization
cols = ["Passes", "Passes Completed", "Attempted Line Breaks", "Completed Line Breaks"]
````

````python

# 4. Apply group-wise z_normalization using groupby.apply
# groupby.apply receives each group's DataFrame; we need to pass cols as well
z_applied = fifa_df.groupby("Team", group_keys=False).apply(lambda g: z_normalization(g, cols))
````

````python
# z_applied now contains the original columns + new columns 'Passes_z', 'Passes Completed_z', etc.
# Example: show first few rows for a team
print(z_applied[z_applied['Team'] == z_applied['Team'].unique()[0]].head())
````

````python
# 5. Now compute the aggregated statistics per team as requested:
# - min "Passes"
# - max "Passes"
# - min "Yellow Cards"
# - max "Yellow Cards"
# - avg "Yellow Cards"
# - max "Attempted Line Breaks"
# - min "Attempted Line Breaks"
# - std of "Attempted Line Breaks"
# - average "Possession"

agg_dict = {
    "Passes_min": ("Passes", "min"),
    "Passes_max": ("Passes", "max"),
    "YellowCards_min": ("Yellow Cards", "min"),
    "YellowCards_max": ("Yellow Cards", "max"),
    "YellowCards_avg": ("Yellow Cards", "mean"),
    "AttemptedLB_max": ("Attempted Line Breaks", "max"),
    "AttemptedLB_min": ("Attempted Line Breaks", "min"),
    "AttemptedLB_std": ("Attempted Line Breaks", lambda x: x.std(ddof=0)),  # ddof=0 for population std
    "Possession_avg": ("Possession (%)", "mean")
}

# Use groupby + agg with named aggregations (pandas >= 0.25 supports this style)
team_stats = fifa_df.groupby("Team").agg(**agg_dict)

# round numeric columns to 2 decimals where appropriate
team_stats = team_stats.round(2)
````

````python

# 6. (Optional) reorder columns to a pleasant ordering
desired_order = [
    "Passes_min", "Passes_max",
    "YellowCards_min", "YellowCards_max", "YellowCards_avg",
    "AttemptedLB_min", "AttemptedLB_max", "AttemptedLB_std",
    "Possession_avg"
]
team_stats = team_stats[desired_order]

# Show results
print(team_stats.head())
print("\n")

# Return both the z-normalized dataframe and aggregated summary
z_all = z_applied  # contains *_z columns
summary = team_stats

# If running interactively, you can display or save these objects.
# e.g.:
display(z_all.head())
print("\n")
display(summary.reset_index().head())

# For convenience in scripts, expose them as variables
z_normalized_df = z_all
team_aggregates_df = summary
````

#### Tips for learning and debugging:

- Print fifa_df.columns to confirm exact column names.
- Use df.head() and df.describe() to understand distributions.
- Use groupby("Team").get_group("YourTeamName").head() to inspect one group.
- If NaNs appear, check with .isna().sum() and consider dropna() or fillna() depending on context.
- Test your z_normalization on a small synthetic DataFrame first to be sure it behaves as expected.

````python
def z_normalization(group, cols_to_perform):
    for col in cols_to_perform:
        std = group[col].std()
        mean = group[col].mean()
        group[f"{col}_z_norm"] = (group[col] - mean) / std
    return group

cols_to_perform = ["Passes", "Passes Completed", "Attempted Line Breaks", "Completed Line Breaks"]


groups = (
    fifa_df
    .groupby("Team", group_keys=False)
    .apply(z_normalization, cols_to_perform=cols_to_perform)
    .groupby("Team")
)

# Alternative step of above 
#  groups = groups.apply(z_normalization, cols_to_perform = cols_to_perform).groupby("Team")

# Aggregation step
result = groups.agg({
    "Passes": ["min", "max"],
    "Yellow Cards": ["min", "max", "mean"],
    "Attempted Line Breaks": ["max", "min", "std"],
    "Possession (%)": "mean"
})

result.head()
````

````python
# Part 02
temp_df = df[~(df.extra_type == "wides")]
temp_df = temp_df.groupby("batter").agg(
{
    "batsman_run": "sum",
    "ballnumber": "count"
})
temp_df["strike_rate"] = temp_df["batsman_run"] / temp_df["ballnumber"] * 100
temp_df[temp_df["ballnumber"] >= 100].sort_values("strike_rate", ascending = False).reset_index().head()
````

##### `Q-7`: Most successful batting pair in IPL. Batting pair who have scored most runs playing together.


````python
# code here
def func(x):
    return '-'.join(list(np.sort(x.values)))
data["batter-pair"] = data[["batter", "non-striker"]].apply(func, axis = 1)
data.groupby("batter-pair")["total_run"].sum().sort_values(ascending = False).head()
````

#### `Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
# code here

temp_df = data.groupby("batter-pair").agg(
{
    "total_run": "sum",
    "ballnumber": "count",
    "isWicketDelivery": "sum"
}).reset_index()

temp_df["Batsman 1"] = temp_df["batter-pair"].apply(lambda x: x.split("-")[0])
temp_df["Batsman 2"] = temp_df["batter-pair"].apply(lambda x: x.split("-")[1])
temp_df.rename(columns = {"total_run": "Runs"}, inplace=True)
temp_df["StrikeRate"] = temp_df["Runs"] / temp_df["ballnumber"] * 100
temp_df["Avg"] = temp_df["Runs"] / temp_df["isWicketDelivery"]
temp_df.sort_values("Runs", ascending = False, inplace=  True)
temp_df[["Batsman 1", "Batsman 2", "Runs", "Avg", "StrikeRate"]]
````
